```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/fraudTrain.csv')
```

```python
legit = df[df.is_fraud == 0]
fraud = df[df.is_fraud == 1]
```

```python
df.drop(columns=['trans_date_trans_time','merchant','category','first','last','gender','street','city','state','job','dob','trans_num'],inplace=True)
```

```python
df.groupby('is_fraud').mean()
```

```python
legit_sample = legit.sample(n=2202)
```

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
 new_dataset.drop(columns=['trans_date_trans_time','merchant','category','first','last','gender','street','city','state','job','dob','trans_num'],inplace=True)
```

```python
new_dataset.groupby('is_fraud').mean()
```

```python
X=new_dataset.drop(columns='is_fraud', axis=1)
Y=new_dataset['is_fraud']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape ,X_test.shape)
```

Model Training

Logistic Regression


```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy)
```

DecisionTreeClassifier

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train, Y_train)
```

```python
X_train_prediction = dtc.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy Score of Model by Decision Tree Classifier : ',training_data_accuracy)
```
